```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data1.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age','affordibility']], df.bought_insurance, test_size=0.2)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

```python
model = keras.Sequential([
    keras.layers.Dense(
        1,
        input_shape=(2,),
        activation="sigmoid"
    )
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train, y_train, epochs=500)
```

```python
model.evaluate(X_test_scaled,y_test)
```

```python
model.predict(X_test_scaled)
```

```python
coef, intercept = model.get_weights()
coef, intercept
```

```python
def sigmoid(cnt):
    import math
    return 1 / (1 + math.exp(-cnt))
sigmoid(71)
```

```python
def predictionFunction(age, affordable):
    weights = coef[0] * age + coef[1] * affordable + intercept
    return sigmoid(weights)
predictionFunction(0.47,1)
```
